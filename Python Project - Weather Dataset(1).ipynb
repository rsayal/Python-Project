### Q6. How many times was the weather foggy in this dataset. 



groupby function in pandas groups the data points (i.e. rows) based on the distinct values in the given column.

```python
#Another way to solve this problem is by filtering the weather column in this dataframe 
#so that only values with the word 'Fog' appears in the result. 

df[df['Weather'] == 'Fog']

```

The above code will work only for the word 'Fog'. However, Python is case sensitive, So, the code as written will only
filter rows where the 'Weather' column value is only the word 'Fog' with a capital 'F'. 
If the values in the 'Weather' column are mixed-case (e.g., 'Fog', 'fog', 'FOG'), this code won't capture all the variations. 
If you want to filter for rows where the 'Weather' column contains the word 'fog' 
regardless of case, you can use the .str.contains() method along with the case parameter set to False:



```python
df[df['Weather'].str.contains('fog', case=False)]
```

## Checking for Statistical Trends with Visual Representation




#### Examine Average Monthly Temperatures over the year 2012 using the Matplotlib Library




In order to calculate the average monthly temperature during the year 2012, first
convert Date/Time column from Object to Datetime datatype

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

```python
dtypes = df.dtypes.reset_index()


# Add column names 
dtypes.columns = ['Column Name', 'Data Type']

dtypes
```

Calculate the monthly average temperature in 2012 using the groupby function along with the .dt.to_period method.The .dt.to_period method converts the datetime column into a specific period or frequency. Ensure that the datetime column is in a proper datetime format using pd.to_datetime(). DateTime Format is more suitable for date-based calculations. The groupby function, combined with the .dt.to_period method, enables the grouping of temperature values by month and the calculation of the mean.  

```python
#Ensure datetime column is in proper datetime format before using groupby and dt.to.period method
```

```python
df.dtypes
```

```python
#Use groupby function and dt.to.period function to calculate the mean temperature per month
```

```python
monthly_avg = df.groupby(df['Date/Time'].dt.to_period('M'))['Temp_C'].mean()
```

```python
#Convert monthly_avg variable from series to dataframe in order to add column names

monthly_avg = monthly_avg.reset_index()


# Add column names to the monthly_avg DataFrame
monthly_avg.columns = ['Year-Month', 'Average temperature Per Month']
```

```python
monthly_avg
```

```python
monthly_avg.dtypes
```

DataType for the year-month column in the monthly average dataframe was period[M] which cant be plotted onto a line graph using 
matplotlib. In order to use this column, convert the datatype from period[M] to timestamp using the dt.to_timestamp method. 
This method converts Period objects to corresponding timestamps, which can then be used to create line plots with matplotlib. For the line graph below, we are using the Year-Month column and the Average Temperatures per month column seen above. 

```python
x = monthly_avg['Year-Month'].dt.to_timestamp()  # Convert Period to Timestamp
y = monthly_avg['Average temperature Per Month']

plt.plot(x, y)
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.xlabel('Year-Month')
plt.ylabel('Average Temperature')
plt.title('Monthly Average Temperature in 2012')
plt.tight_layout()
plt.show()
```
